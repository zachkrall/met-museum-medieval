```python
import requests
import json
import time
import os
from pathlib import Path
import logging
from typing import List, Dict
import pandas as pd
import csv
```

```python
df = pd.read_csv('met_objects.csv')
```

```python
# for each row in the dataframe
for index, row in df.iterrows():
    id = row['objectID']
    url = row['primaryImage']

    if pd.isnull(url):
        continue

    # check if the image exists in local images folder `{id}.jpg`
    if not os.path.exists(f'images/{id}.jpg'):
        # if not, download the image
        response = requests.get(url)
        if response.status_code == 200:
            with open(f'images/{id}.jpg', 'wb') as f:
                f.write(response.content)
            print(f"Downloaded {id}.jpg")
        else:
            print(f"Failed to download {id}.jpg")
    else:
        print(f"Already downloaded {id}.jpg")
```
